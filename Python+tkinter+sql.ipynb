```python
from tkinter import *
import pymysql # importing sql yog
import tkinter.messagebox as me


root= Tk() # creating main window
root.geometry('500x500') # define size of main window
root.title('My Window')
root.configure(bg='sky blue') # define background color

# connecting to sql
def createconn():
    return pymysql.connect(host='localhost',database='MyDemo1',
                           user='root',password='2580',port=3306)

def insertdata():
    r=ern.get()
    f=efn.get()
    l=eln.get()
    em=eem.get()
    mo=emo.get()
    
    if (r=="" or f=="" or l=="" or em=="" or mo==""):
        me.showinfo('Insert status','All entry is mandatory')
        
    else:
        try:
            conn=createconn()
            cursor=conn.cursor()
            args=(r,f,l,em,mo)
            query="insert into students(sid,fname,lname,email,mobile) values (%s,%s,%s,%s,%s)"
            cursor.execute(query,args)
            conn.commit()
            me.showinfo('Insert status','Data inserted')
            conn.close()
        except Exception as ee:
            me.showinfo('Insert status','Error')
            print('Insert Exception: ',ee)

# creating labels

rn=Label(root,text='Roll no.')
rn.place(x=150,y=40)

fn=Label(root,text='First name')
fn.place(x=150,y=80)

ln=Label(root,text='Last name')
ln.place(x=150,y=120)

em=Label(root,text='Email ID')
em.place(x=150,y=160)

mo=Label(root,text='Mobile no.')
mo.place(x=150,y=200)


# user input

ern=Entry()
ern.place(x=250,y=40)

efn=Entry()
efn.place(x=250,y=80)

eln=Entry()
eln.place(x=250,y=120)

eem=Entry()
eem.place(x=250,y=160)

emo=Entry()
emo.place(x=250,y=200)

# creating button

button1 = Button(root,text="SAVE",command=insertdata)
button1.place(x=200,y=250)

mainloop()
```
